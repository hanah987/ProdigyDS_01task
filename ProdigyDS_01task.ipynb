```python
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Iterate over the numeric columns and plot histograms
for col in numeric_cols:
    fig = plt.figure(figsize=(5,5))
    plt.hist(df[col], bins=10, edgecolor='black')
    plt.xlabel(col)
    plt.title(f'Histogram of {col}')
    plt.show()
    print("\n")
```

```python
import matplotlib.pyplot as plt

# Select only columns that contain population data for each year (numeric columns)
# Assuming columns related to population data are years (e.g., '1960', '1961', ..., '2020')
year_columns = df.columns[1:]  # Exclude the first column, which is likely 'Country Name'

# Loop through each year and plot a histogram
for year in year_columns:
    fig = plt.figure(figsize=(8,5))
    plt.hist(df[year].dropna(), bins=10, color='skyblue', edgecolor='black')  # Drop NaN values for plotting
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Population in {year}')
    plt.show()
    print("\n")
```

```python
import matplotlib.pyplot as plt

# Select only columns that contain population data for each year (numeric columns)
year_columns = df.columns[1:]  # Exclude the first column, which is likely 'Country Name'

# Loop through each year and plot a line plot
for year in year_columns:
    fig = plt.figure(figsize=(8,5))
    plt.plot(df['Country Name'], df[year], marker='o', linestyle='-', color='skyblue', markerfacecolor='black')  # Plot population vs. country name
    plt.xlabel('Country Name')
    plt.ylabel('Population')
    plt.title(f'Population in {year}')
    plt.xticks(rotation=90)  # Rotate x-axis labels if there are many countries
    plt.grid(True)
    plt.show()
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # For density plots

# Assuming df is your DataFrame and the first column is 'Country Name'
year_columns = df.columns[1:]  # Exclude the first column, which is 'Country Name'

# Filter year columns from 1961 to 2021
year_columns = [year for year in year_columns if year.isdigit() and 1961 <= int(year) <= 2021]

# Line plot
for year in year_columns:
    fig = plt.figure(figsize=(8, 5))
    plt.plot(df[year].dropna(), color='red')  # Drop NaN values for plotting
    plt.xlabel('Index')
    plt.ylabel('Population')
    plt.title(f'Line Plot of Population in {year}')
    plt.show()
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # For density plots

# Filter year columns to include only from 1961 to 2021
year_columns = [str(year) for year in range(1961, 2022)]

# Ensure year_columns match the DataFrame columns
year_columns = [col for col in year_columns if col in df.columns]

# Define colors for alternating lines
colors = ['red', 'yellow']

# Line plot
for year in year_columns:
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot data for each year in alternating colors
    for i, year in enumerate(year_columns):
        color = colors[i % len(colors)]  # Alternate colors
        ax.plot(df.index, df[year].dropna(), color=color, label=year)

    ax.set_xlabel('Index')
    ax.set_ylabel('Population')
    ax.set_title('Line Plot of Population from 1961 to 2021')
    ax.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)  # Rotate index labels for better readability
    plt.tight_layout()       # Adjust layout to fit labels
    plt.show()
    print("\n")
```
